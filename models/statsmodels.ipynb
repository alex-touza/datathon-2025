## Statsmodels
This model is lighter and works pretty fast, but only reaches a f1

```python

import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.metrics import f1_score


DATASET_PATH = '../data/dataset.csv'

features: list[str] = []

features_data_list: list[list[float]] = []

df = pd.read_csv(DATASET_PATH)

target_data = df['target_variable'].astype(int)

feature_cols = [
    "product_A_sold_in_the_past",
    "product_B_sold_in_the_past",
    "product_A_recommended",
    "product_A",
    "product_C",
    "product_D",
    "cust_hitrate",
    "cust_interactions",
    "cust_contracts",
    "opp_month",
    "opp_old",
    "competitor_Z",
    "competitor_X",
    "competitor_Y",
    "cust_in_iberia",
]
dataset = df[feature_cols + ['target_variable']]

# Split dataset in train and test
train_data, test_data = np.split(dataset.sample(frac=1, random_state=42), [int(0.8 * len(dataset))])

# Extract target variables for train and test
train_target = train_data['target_variable'].astype(int)
test_data_corr = test_data['target_variable'].astype(int)

# Drop target variable from feature sets
train_data = train_data[feature_cols]
test_data = test_data[feature_cols]

X = sm.add_constant(train_data)

model = sm.Logit(train_target, X)
result = model.fit()

print(result.summary())
```
