```python
import numpy as np
import statsmodels.api as sm

# Example data: X has two predictors, y is binary
X = np.array([[0.2, 1.3],
              [1.1, 0.7],
              [0.4, 2.2],
              [2.0, 0.1],
              [1.3, 1.5]])

y = np.array([0, 0, 0, 1, 1])

# Add intercept
X = sm.add_constant(X)

model = sm.Logit(y, X) # Genera els pesos
result = model.fit()

print(result.summary())

new_obs = np.array([1.0, 0.8, 1.6])  # [constant, x1, x2]
pred_prob = result.predict(new_obs)
print(pred_prob)
```
